```python
#1
listaname=[]
listaage=[]
name=str(input("Ingrese el nombre: "))
age=int(input("Ingrese la edad: "))
listaname.append(name)
listaage.append(age)
decision=str(input("¿Desea añadir mas datos?: "))
while decision=="si" or decision=="Si" or decision=="sí" or decision=="Sí":
  name=str(input("Ingrese el nombre: "))
  age=int(input("Ingrese la edad: "))
  listaname.append(name)
  listaage.append(age)
  decision=str(input("¿Desea añadir mas datos?: "))
print("//////////////////")
for x,y in zip(listaname,listaage):
  print(x,y)
```

```python
#1.2
import matplotlib.pyplot as plt
import numpy as np
coo=np.arange(len(listaage))
anc=0.30
fig, ax=plt.subplots()
#ax.bar(Coordenadas,Altura,Ancho)
ax.bar(coo,listaage,anc)
ax.set_title("Grafico de barras")
ax.set_ylabel("Edad")
ax.set_xlabel("Nombre")
ax.set_xticks(coo)
ax.set_xticklabels(listaname)
for i,j in zip(coo,listaage):
  ax.annotate(j, xy=(i-0.05,j+0.2))
```

```python
#2
listax=[]
listay=[]
a=-100
b=100
c=-100
while c>=a and c<=b:
  d=c**3
  listax.append(c)
  listay.append(d)
  c=c+1
for x,y in zip(listax,listay):
  print(x,y)
```

```python
#2.2
import matplotlib.pyplot as plt
import numpy as np
coor=np.arange(len(listay))
an=0.35
gf, ax=plt.subplots(figsize=(50,30))
ax.bar(coor,listay,an)
ax.set_xticks(coor)
ax.set_xticklabels(listax)
ax.set_title("Grafico de barras")
ax.set_xlabel("X")
ax.set_ylabel("X**3")
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np
from random import randint
listaA=[]
listaB=[]
a=1
b=20
c=1
while c>=a and c<=b:
  d=randint(1,6)
  listaA.append(c)
  listaB.append(d)
  c=c+1
for x,y in zip(listaA,listaB):
  print(x,y)
```

```python
#3.2
import matplotlib.pyplot as plt
import numpy as np
anch=0.35
coord=np.arange(len(listaB))
gf, ax=plt.subplots()
ax.bar(coord,listaB,anch)
ax.set_xticks(coord)
ax.set_xticklabels(listaA)
ax.set_title("Grafico de barras")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
#3.3
import numpy as np
listaC=listaA+listaB
listaC=np.array(listaC)
print("La media es: ",np.mean(listaC))
print("La mediana es: ",np.median(listaC))
print("La desviación estandar en Y es de: ",np.std(listaB))
```

```python
#5
import numpy as np
n=int(input("Ingrese el valor de n: "))
Matriz=np.random.randint(10,99,size=(n,n))
print(Matriz)
```

```python
#5.2
mp=[]
ms=[]
import matplotlib.pyplot as plt
import numpy as np
for i in range(n):
  #Primer valor renglón. Segundo valor columna.
  mp.append(Matriz[i][i])
  ms.append(Matriz[i][n-i-1])
print("Diagonal principal: ",mp)
print("Diagonal secundaria: ",ms)
gf, ax=plt.subplots()
plt.scatter(mp,ms)
ax.set_xticks(mp)
ax.set_title("Elementos de las diagonales")
ax.set_xlabel("Diagonal principal")
ax.set_ylabel("Diagonal secundaria")
```
