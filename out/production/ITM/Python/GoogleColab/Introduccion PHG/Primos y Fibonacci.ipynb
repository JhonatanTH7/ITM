```python
#versión1
def esPrimo1():
  num=int(input("Ingrese el número entero: "))
  if num==1:
    print("No es Primo")
  else:
    cont=2
    for i in range(2,num):
      if num%i==0:
        cont=cont+1
    if cont>2:
      print("No es Primo")
    else:
      print("SI es Primo")
```

```python
esPrimo1()
```

```python
#versión2
def esPrimo2(num):
  #num=int(input("Ingrese el número entero: "))
  if num==1:
    return "No es Primo"
  else:
    cont=2
    for i in range(2,num//2 + 1):
      if num%i==0:
        cont=cont+1
    if cont>2:
      return "No es Primo"
    else:
      return "SI es Primo"
```

```python
print(esPrimo2(1000000000000000))
```

```python
#versión3
import math
def esPrimo3(num):
  if num==1:
    return "No es Primo"
  else:
    cont=2
    for i in range(2,int(math.sqrt(num))+ 1):
      if num%i==0:
        cont=cont+1
    if cont>2:
      return "No es Primo"
    else:
      return "SI es Primo"
```

```python
print(esPrimo3(19))
```

```python
#versión4
import math
def esPrimo4(num):
  if num==1:
    return "No es Primo"
  else:
    cont=2
    i=2
    while i < int(math.sqrt(num))+ 1:
      if num%i==0:
        cont=cont+1
        i=num+1 #obliga a que la condición del mientras sea falsa
      i=i+1
    if cont>2:
      return "No es Primo"
    else:
      return "SI es Primo"
```

```python
print(esPrimo4(100000001))
```

```python
def sumaFibinacci(n):
  if n==1:
    s=0
  else:
    if n==2:
      s=1
    else:
      t1=0
      t2=1
      s=1
      fibo=1
      for i in range(3,n+1):
        fibo=t1+t2
        s=s+fibo
        t1=t2
        t2=fibo
  return s
```

```python
print(sumaFibinacci(1))
```
