#Enunciado
Elaborar un programa en phyton que permita ingresar los datos para n ventas. de cada venta se conoce: codigo venta.número de unidades vendidas y valor por unidad. Calcular y mostrar

1.Promediio de ventas

2.Código y valor de la venta mayor

3.Código y valor de la venta menor

4.Porcentaje de ventas mayores que el promedio

```python
#Solución en phyton
#Inicializar contadores, acumuladores y otras variables
sumaVentas=0
contMayProm=0
mayVenta=0
codMay=""
menVenta=100000000000000
codMen=""

#Datos de entrada
n=int(input("Ingrese el número de ventas"))
for i in range(1,n+1): #i comienza en 1 y termina en n
  codVenta=input("Ingrese el código de la venta#"+str(i)+": ")
  numUnid=int(input("Ingrese el número de unidades vendidas en la venta#"+str(i)+": "))
  valorUnid=float(input("Ingrese el valor por unidad en la venta#"+str(i)+": "))
  venta=valorUnid*numUnid
  sumaVentas=sumaVentas+venta
  if venta > mayVenta:
    mayVenta=venta
    codMay=codVenta
  if venta < menVenta:
    menVenta=venta
    codMen=codVenta

prom=sumaVentas/n
for i in range(1,n+1): #i comienza en 1 y termina en n
  numUnid=int(input("Ingrese el número de unidades vendidas"))
  valorUnid=float(input("Ingrese el valor por unidad"))
  venta=valorUnid*numUnid
  if venta > prom:
    contMayProm=contMayProm+1

porc=contMayProm*100/n

print(prom)
print(codMay,mayVenta)
print(codMen,menVenta)
```
