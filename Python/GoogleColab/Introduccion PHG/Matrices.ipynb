```python
for i in range(len(mat1)):
  for j in range(len(mat1[0])):
    print(mat1[i][j],end="\t")
  print()
```

```python
#primero se debe llenar de ceros para asignar un espacio de memoria, en caso de no hacerlo tira error
mat3=[]
nfilas=5
ncolum=5
for i in range(nfilas):
  mat3.append([])
  for j in range(ncolum):
    mat3[i].append(0)
print(mat3)

import random
for i in range(nfilas):
  for j in range(ncolum):
    mat3[i][j]=random.randint(1,9)
print(mat3)
```
